```python
df = pd.read_csv("../data/train_t1d.data")
df1 = df.loc[df["T1D"] == True, :]
df1.drop("Unnamed: 0", 1, inplace=True)
df1.drop("TRA", 1, inplace=True)
df1.drop("T1D", 1, inplace=True)
df1.columns = ["aminoAcid"]
df1["counts"] = 1
df1["v_beta"] = ""
df1["j_beta"] = ""
df1.to_csv("../../DeepTCR/Data/T1D/Positive/Positive.tsv", sep="\t", index=None)
df1
```

```python
df = pd.read_csv("../data/train_t1d.data")
df1 = df.loc[df["T1D"] == False, :]
df1.drop("Unnamed: 0", 1, inplace=True)
df1.drop("TRA", 1, inplace=True)
df1.drop("T1D", 1, inplace=True)
df1.columns = ["aminoAcid"]
df1["counts"] = 1
df1["v_beta"] = ""
df1["j_beta"] = ""
df1.to_csv("../../DeepTCR/Data/T1D/Negative/Negative.tsv", sep="\t", index=None)
df1
```
